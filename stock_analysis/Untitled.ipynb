```python
import requests
import pandas as pd

def crawl_Dividend_Report(url, stockNo):
    
    form_data = {
        'encodeURIComponent':1,
        'step':1,
        'firstin':1,
        'off':1,
        'co_id':stockNo,
        'TYPEK':"all"
    }

    r = requests.post(url,form_data)
    html_df = pd.read_html(r.text)[1].fillna("")
    return html_df

a = crawl_Dividend_Report("https://mops.twse.com.tw/mops/web/ajax_t05st09", 2812)
a.head(20)
```

```python
import pandas as pd

url = 'https://tw.stock.yahoo.com/d/s/dividend_2412.html'
# Encode this website by Big5
df = pd.read_html(url,encoding="Big5")
# Delete first line, delete chinese columns
df = df.drop(0).reset_index(drop=True) 
# Name columns
df.columns = ['year','CASH','SRE','SCAP','STOCK','TOTAL']
# save to .csv
df.to_csv(r'D://dividend/2412.csv',index=False,sep=',')
```
